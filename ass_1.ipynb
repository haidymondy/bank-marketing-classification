```python
import pandas as pd
train=pd.read_csv("train.csv")
"""train.head()
train.info()"""
```

```python
train.describe().T
```

```python
from sklearn.preprocessing import LabelEncoder
col=['job','marital','education','default','housing','loan','month','poutcome','contact','day_of_week',]
for c in col:
    lbl=LabelEncoder()
    train[c]=lbl.fit_transform(train[c].values)


train.info()
```

```python
y = train["y"]
X = train.drop(["y"], axis = 1)

lab = LabelEncoder()
y=lab.fit_transform(y)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split, GridSearchCV

pipeline = Pipeline([
    ('min_max_scaler', MinMaxScaler()),
    ('std_scaler', StandardScaler())])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle = True, random_state=0)
X_train = pipeline.fit_transform(X_train)
X_test = pipeline.transform(X_test)
```

```python
import numpy as np
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import RMSprop
from keras.layers import Dense
import matplotlib.pyplot as plt
import tensorflow as tf

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.model_selection import GroupShuffleSplit

from tensorflow import keras

from tensorflow.keras import layers
from tensorflow.keras import callbacks
```

```python
X_train.shape
```

```python
batch_size = 150
#num_classes = 1
epochs = 250

# Create a sequential model here
model = Sequential()

#model.add(Dense(200, activation='relu', input_shape=X_train.shape[1:]))
model.add(Dense(128, activation='relu', input_dim=20))

layers.BatchNormalization()
layers.Dropout(0.3)


model.add(Dense(64, activation='relu'))

layers.BatchNormalization()
layers.Dropout(0.3)

model.add(Dense(32, activation='relu'))

layers.BatchNormalization()
layers.Dropout(0.3)

model.add(Dense(16, activation='relu'))

layers.BatchNormalization()
layers.Dropout(0.3)

model.add(Dense(8, activation='relu'))

layers.BatchNormalization()
layers.Dropout(0.3)


model.add(Dense(1, activation='sigmoid'))

model.summary()


```

```python
"""model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['accuracy'])"""

model.compile(optimizer='adam',
    loss='binary_crossentropy',
    metrics=['binary_accuracy'],)

#model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

history = model.fit(X_train, y_train,
                    batch_size=batch_size,
                    epochs=epochs,
                    verbose=1,
                    validation_split=0.2)
```

```python
score = model.evaluate(X_test, y_test, verbose = 1)
print('Test score:', score[0])
print('Test accuracy:', score[1])
#8968
```
